**Vader Sentiment Scoring**

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores("I'am happy")
```

```python
sia.polarity_scores(example)
```

```python
df.head()
```

```python
res = {}
for i , row in tqdm(df.iterrows(),total = len(df)):
  try:
    text = row['Text']
    myid = row['id']
    # Check if the text is a string before processing
    if isinstance(text, str):
      res[myid] = sia.polarity_scores(text)
  except RuntimeError:
    print(f"broke at {myid}")
```

```python
vaders = pd.DataFrame(res).T
vaders.head()
```

```python
vaders = vaders.reset_index().rename(columns = {'index' : 'id'})
vaders = vaders.merge(df,how = 'left')
```

```python
import seaborn as sns
ax = sns.barplot(data = vaders,x = 'Score',y = 'compound')
ax.set_title('compound score by amazon star review')
plt.show()
```

```python
fig,axs = plt.subplots(1,3,figsize = (15,3))
sns.barplot(data = vaders, x = 'Score' , y = 'neg' ,ax = axs[0])
sns.barplot(data = vaders, x = 'Score' , y = 'neu' ,ax = axs[1])
sns.barplot(data = vaders, x = 'Score' , y = 'pos' ,ax = axs[2])
axs[0].set_title('Negative')
axs[1].set_title('Neutral')
axs[2].set_title('Positive')
plt.show()
```

**Roberta Pretrained Model**

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
from transformers import pipeline
```

```python

MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
encoded_text = tokenizer(example,return_tensors = 'pt')
output = model(**encoded_text)
output
```

```python
scores = output[0][0].detach().numpy()
scores = softmax(scores)
scores
```

```python
scores_dict = {'roberta_neg' : scores[0],'roberta_neu' : scores[1],'roberta_pos' : scores[2]}
```

```python
print(scores_dict)
```

```python
def polrity_scores(example):
  encoded_text = tokenizer(example,return_tensors = 'pt')
  output = model(**encoded_text)
  scores = output[0][0].detach().numpy()
  scores = softmax(scores)
  scores_dict = {'roberta_neg' : scores[0],'roberta_neu' : scores[1],'roberta_pos' : scores[2]}
  return scores_dict
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
res = {}
for i , row in tqdm(df.iterrows(),total = len(df)):
  try:
    myid = row['id']
    text = row['Text']
    vader_result = sia.polarity_scores(text)
    roberta_result = polrity_scores(text)
  except RuntimeError:
      continue
```

```python
both = {**vader_result,**roberta_result}
res[myid] = both
```

```python
result_df = pd.DataFrame(res).T
result_df = result_df.reset_index().rename(columns = {'index' : 'id'})
result_df = result_df.merge(df,how = 'left')
```
